```python
# import Pandas
import pandas as pd

# declare a Series
my_series = pd.Series([1, 2, 3])
print(my_series)

# print value by index
print(my_series[0])

# add labelled indexes
my_series.index = ["x", "y", "z"]
print(my_series)

# print value by index to see warning
print(my_series[0])

# print value by index with iloc[]
print(my_series.iloc[0])

# print value by label
print(my_series["x"])

# declare new series with labels using kwargs index=
my_series = pd.Series([4, 5, 6], index=["a", "b", "c"])
print(my_series)

# declare new series with labels from dictionary
my_series = pd.Series({ "d": 7, "e": 8, "f": 9 })
print(my_series)
```

```python
pets.shape
pets.head()
pets.tail(3)
pets.dtypes
pets.info()
pets.describe()
```

```python
# sort by name, then animal/name and animal/age
pets.sort_values(by=["animal", "age"])

# create duplicate, then check for duplicate with & without subset
pets.duplicated(subset=["name", "animal"])

# check for uniques and nulls
pets.nunique()
pets.isna()

# value count on df, then animal column
pets.value_counts()
pets["animal"].value_counts()
```

```python
# multiple ways to target label to be renamed
pets.rename(columns={ "name": "animal_name"})
pets.rename({ "name": "animal_name" }, axis=1)

# drop without inplace=True doesn't affect original df
pets.drop(["name"], axis=1)

# create null value to drop
pets.iloc[1,2] = None

# drop null rows with null values
dropped = pets.dropna()
print(dropped)

# reset the indexes
dropped.reset_index(inplace=True)
dropped
```

```python
# demo df.mean() throws error because of string values
# pets.mean()

# drop non-numeric columns and retry
pets.drop(["name", "animal", "desexed"], axis=1).mean()

# apply to a single column
pets["age"].mean()

# filter a column then apply
pets["age"][pets["animal"] == "dog"].mean()

# groupby animal, then get mean for single column
pets.groupby("animal")["age"].mean()
```

```python
# define a function that takes the whole row as an argument (print row to start)
# extract animal from row and return sound
def what_does_the_animal_say(row):
  # print(row)
  if row['animal'] == "cat":
    return "meow"
  if row['animal'] == "dog":
    return "woof"
  if row['animal'] == "bird":
    return "tweet"
  return "???"

# apply that function to df
pets.apply(what_does_the_animal_say, axis=1)

# apply lambda function that does the same thing to single column
# create new column for sound
pets["sound"] = pets["animal"].apply(
  lambda animal : "woof" if animal == "dog" else "meow" if animal == "cat" else "tweet"
)
pets
```
