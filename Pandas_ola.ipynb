```python
import pandas as pd
import numpy as np
movies = pd.read_csv("movies_2.csv")
#movies.shape
#movies.info()
columns_to_check = ["MPAA Rating", "Budget", "Gross", "Release Date", "Genre", "Runtime", "Rating", "Rating Count"]
nan_count = movies[columns_to_check].isna()
nan_count.sum()
```

```python
#movies = movies.drop(columns=["Summary"])
movies = movies.rename(columns=str.lower)
movies = movies.drop(movies.index[-1])   # removing indeks 615 will not remove always the last indeks
#movies.rename(columns = lambda x: x.replace(' ','_'), inplace = True)
movies.columns = movies.columns.str.replace(" ","_") #overwriting a column with replaced value

how_many = movies["rating"].isna().value_counts()

movies.fillna({ 'rating': movies['rating'].median() }, inplace=True)
#how_many2 = movies["rating_count"].isna().value_counts()
movies.fillna({ 'rating_count': movies['rating_count'].median() }, inplace=True)
#movies.tail()
#duplicates = movies.duplicated(["titel"]).sum() keep: 2 subset = ["titel"]["release_date"] try it 

movies["budget"]= movies["budget"].astype("int32")
movies["gross"] = movies["gross"].astype("int32")
movies["release_date"] = pd.to_datetime(movies["release_date"])
movies.tail()
```
